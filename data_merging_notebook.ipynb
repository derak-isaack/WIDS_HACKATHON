### Train data conversion to parquet file 

```python
conn1 = duckdb.connect('wids.db')
query = """
        SELECT * 
        FROM read_csv('TRAIN\TRAIN_FUNCTIONAL_CONNECTOME_MATRICES_new_36P_Pearson.csv')
"""

data = duckdb.execute(query).fetch_df()
data = pd.DataFrame(data)
data.head()
```

```python
connect2 = duckdb.connect('wids_train.db')
data2 = connect2.execute("SELECT * FROM read_xlsx('TRAIN/TRAIN_QUANTITATIVE_METADATA_new.xlsx')").fetch_df()
data2 = pd.DataFrame(data2)
data2.head()
```

```python
connect3 = duckdb.connect('wids_train.db')
data3 = connect3.execute("SELECT * FROM read_xlsx('TRAIN/TRAIN_CATEGORICAL_METADATA_new.xlsx')").fetch_df()
data3 = pd.DataFrame(data3)
data3.head()
```

```python
for col in data3.columns[1:]:  
    if pd.api.types.is_float_dtype(data3[col]):
        if (data3[col] % 1 == 0).all():  
            data3[col] = data3[col].astype(int)
```

```python
data3.head()
```

```python
connect4 = duckdb.connect('wids_train.db')
data4 = connect4.execute("SELECT * FROM read_csv('NILEARN_805TRAINING_ADHD.csv')").fetch_df()
data4.drop(columns='column000', axis=1, inplace=True)
data4 = pd.DataFrame(data4)

data4.head()
```

```python
connect5 = duckdb.connect('wids_train.db')
data5 = connect5.execute("SELECT * FROM read_xlsx('TRAIN/TRAINING_SOLUTIONS.xlsx')").fetch_df()
data5 = pd.DataFrame(data5)
data5.head()
```

```python
data5[["ADHD_Outcome", "Sex_F"]] = data5[["ADHD_Outcome", "Sex_F"]].astype(int)
data5.head()
```

```python
query = """
SELECT *
FROM data4
JOIN data5 USING (participant_id)
"""

# Execute the query
result_df = duckdb.query(query).to_df()
result_df
```

```python
query = """
SELECT *
FROM data2
JOIN data3 USING (participant_id)
JOIN data5 USING (participant_id)
"""

# Execute the query
supplimentary_df = duckdb.query(query).to_df()
supplimentary_df
```

```python
duckdb.register('result_table', result_df)
duckdb.query("COPY result_table TO 'nilearn_805FULL_TRAINING.csv' (FORMAT 'csv', HEADER)")
```

```python
duckdb.register('result_table', supplimentary_df)
duckdb.query("COPY result_table TO 'suplimentary_train.csv' (FORMAT 'csv', HEADER)")
```

### **Join the selected categorical columns with the reduced 800 features from PCA.**

```python
connect_cat = duckdb.connect('wids_train.db')
train_cat = connect_cat.execute("SELECT * FROM read_csv('cat_columns.csv')").fetch_df()
train_cat = pd.DataFrame(train_cat)
train_cat.head()
```

```python
connect_reduced = duckdb.connect('wids_train.db')
train_reduced = connect_reduced.execute("SELECT * FROM read_csv('800Reduced_functional_brain.csv')").fetch_df()
train_reduced = pd.DataFrame(train_reduced)
train_reduced.head()
```

```python
#Merge teh reduced daataframe s
reduced_train_query = """
SELECT *
FROM train_cat
JOIN train_reduced USING (participant_id)
"""

# Execute the query
result_df_reduced = duckdb.query(reduced_train_query).to_df()

duckdb.register('train_reduced_table', result_df_reduced)
duckdb.query("COPY train_reduced_table TO '802TRAIN_FUNCTIONAL.csv' (FORMAT 'csv', HEADER)")
```
